```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/Housing.csv")
df
```

```python
df.info()
```

```python
plt.scatter(df.area,df.price,color ='blue',marker = '+')
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Housing Area Vs Price")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
print("Single Variable")
print("Slope = ", reg.coef_)
print("Intercept = ", reg.intercept_)
a =float(input("Enter the Area:"))
sklea = reg.predict([[a]])
print("Predicted Price(skicit) = ", sklea)
prediction = reg.coef_ * a + reg.intercept_
print("Predicted Price(Manually) = ", prediction)
```

```python
plt.scatter(df.area,df.bedrooms,color ='green',marker = '+')
plt.xlabel("Area")
plt.ylabel("Bedrooms")
plt.title("Housing Area Vs Bedrooms")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms']],df.price)
```

```python
print("MultiVariable")
print("Slope = ", reg.coef_[0])
print("Slope = ", reg.coef_[1])
print("Intercept = ", reg.intercept_)
a =float(input("Enter the Area:"))
b =float(input("Enter the Bedrooms:"))
sklea = reg.predict([[a,b]])
print("Predicted Price(skicit) = ", sklea)
abc = reg.coef_[0] * a + reg.coef_[1] * b + reg.intercept_
print("Predicted Price(manual) = ",abc)
```
